```python
# Identify numeric columns
numeric_cols = more_selected.select_dtypes(include=[np.number]).columns

# Calculate Q1, Q3 and IQR only for numeric columns
Q1 = more_selected[numeric_cols].quantile(0.25)
Q3 = more_selected[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

# Filter the DataFrame to remove outliers in numeric columns
more_selected = more_selected[~((more_selected[numeric_cols] < (Q1 - 1.5 * IQR)) | (more_selected[numeric_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```
